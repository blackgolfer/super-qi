## First Pappus circle

**Definition**: Given an arbelos, there is a circle $(\delta)$ included in this arbelos and tangent to $(\alpha)$, $(\beta)$ and $(\gamma)$; it is called inscribed circle or circle of Pappus. Archimedes also determined the radius of this circle.\\
The radius of the Pappus circle is $r=\dfrac{ab(a+b)}{a^2+ab+b^2}$.

Let us first analyze constructions of this circle in the general case, then we will examine the particular case with the golden arbelos.

Given a golden arbelos $(A,C,B)=(\alpha,\beta,\gamma)$, we now that $(\beta)$ and $\gamma$ are external tangent at $C$
  and internal tangent at $(\alpha)$ at $A,B$. Construct the circle $(\delta)$ external tangent at $((\alpha),(\beta),(\gamma))$. Let us note $P_0, P_1, P_2$ the points of tangency.


### Construction of Arbelos with arcs version 1

See paragraph [\ref{sub:points_cocylic}] for a justification and the document ”Simple Constructions of the Incircle of an Arbelos”. 

Without justifications (they will be given later), here is one of the simplest constructions.

- An arbelos is given the semicircles $(\alpha)$, $(\beta)$ and $(\gamma)$ have centers $O_0$, $O_1$ and $O_2$.
  $Q$ and $P$ are the north poles of $(\beta)$ and $(\gamma)$.
- Two arcs of circles of respective centers $P$ and $Q$ passing through $C$ determine on the arbelos the points of tangency of the circle $(\delta)$ with the arbelos: $P_0$, $P_1$ and $P_2$. 
- The center of the Pappus circle is given by the intersection of the lines $(O_1P_1)$ and $(O_2P_2)$.

### Explanation

Let $O_4$ be the center of $(\delta)$ and $H_4$ its orthogonal projected onto the line $(AB)$. $H_4$ belongs to the segment $[CO_2]$ since we have assumed $a>=b$.

From the right-angled triangles $O_0O_4H_4$, $O_1O_4H_4$ and $O_2O_4H_4$ we get

  \begin{align*}
    O_4O_0^2&=O_4H_4^2+O_0H_4^2 \\
    O_4O_1^2&=O_4H_4^2+O_1H_4^2\\
    O_4O_2^2&=O_4H_4^2+O_1H_4^2
  \end{align*}
Either
  \begin{align*}
    (a+b-r)^2&=O_4H_4^2+O_0H_4^2 \\
   (a+r)^2&=O_4H_4^2+(b+O_0H_4)^2\\
    (b+r)^2&=O_4H_4^2+(a-O_0H_4)^2
  \end{align*}

because

$O_1H_4=O_1O_0+O_0H_4=O_0H_4+(AO_0−AO_1)=O_0H_4+(a+b)−a=b+O_0H_4$

and

$O_2H_4=O_0O_2−O_0H_4=O_1O_2-O_0O_2 -O_0H=(a+b)−b−O_0H_4=a−O_0H_4$.

From the first two equalities we get

$b\cdot O_0H_4=2ar-b^2-ab+br$

and from the first and third we get

$a\cdot O_0H_4=a^2+ab-ar-2br$. Hence
$r=\dfrac{ab(a+b)}{a^2+ab+b^2}$.

[The Book of Lemmas: Proposition 1 What is this about? A Mathematical Droodle](https://www.cut-the-knot.org/Curriculum/Geometry/BookOfLemmas/BOL1.shtml)

Let $[GH]$ be the diameter of the circle which is parallel to $(AC)$, and let the circle touch the semicircles on $[AC]$, $[AB]$, $[BC]$ in $P_1$, $P_0$, $P_2$ respectively.

Then, by Prop. 1 $A$, $G$ and $P_0$ are aligned, as well as $P_0$, $H$ and $B$.\\
 For a like reason $A$ $P_1$ and $H$ are aligned, $B$ $P_2$ and $G$ are aligned, as also are $C$ $P_1$ and $G$, $C$ $P_2$ and $H$.
 
Let $(AP_0)$ meet the semicircle on $(\beta)$ at $I$, and let $(BP_0)$ meet the semicircle on $(\gamma)$ in $J$. Join $(CI)$, $(CJ)$ meeting $(AP_1)$, $(BP_2)$ in $K$, $L$, and let $(GL)$, $(HM)$ produced meet $(AB)$ in $F$, $E$ respectively.

Now, in the triangle $AGB$, the perpendiculars from $A$, $C$ on the opposite sides meet in $K$. Therefore by the properties of triangles, $(GF)$ is perpendicular to $(AC)$.
Similarly $(HE)$ is perpendicular to $(BC)$.\\
The polygon $EFGH$ is a rectangle. Let's show that it is a square. 

- It is easy to show that the lines $(CJ)$ and $(AP_0)$ are parallel as are the lines $(BP_0)$ and $(CI)$.

  Therefore $\dfrac{AC}{BC} = \dfrac{AL}{LH} =\dfrac{AF}{FE} $, and $\dfrac{BC}{AB} = \dfrac{CM}{MG} =\dfrac{BE}{EF} $.

  Hence  $\dfrac{AF}{EF} = \dfrac{EF}{BE}$ (1). (All this is used in [proposition 6 of the book of lemmas](https://www.cut-the-knot.org/Curriculum/Geometry/BookOfLemmas/BOL6.shtml).)

- Now we can notice that the right triangles $AGF$ and $EBH$ are similar. Thus $\dfrac{GF}{AF} = \dfrac{BE}{HE} = \dfrac{BE}{GF}$ (2).
- With (1) and (2) $GF^2=EF^2$ so $GF=EF$. The rectangle $EFGH$ is a square.


Now suppose that $B$ divides $[AC]$ such as :

$$k =  \frac{AC}{CB} \quad\text{then}  \quad AF = kEF \quad \text{and} \quad EF=  kBE $$

We have 

$$ AB = AF + FE + EB\quad \text{either} \quad AC + CB = AF+ FE + EB $$

 or 
 
$$AB = kFE + FE + EF/k = \frac{k^2+k+1}{k}EF $$

Finally

$$ \frac{AB}{GH} =  \frac{k^2+k+1}{k} $$

Conclusion: $EFGH$ is a square with $[GH]$ diameter of the pappus circle. $A$, $P_1$ and $H$ are aligned, as well as $A$, $G$ $P_0$ and $B$,$ P_2$, $G$.

### Points cocylic

**Lemma (Points coclic)**: Let $P$ and $Q$ be the midpoints of the semicircles $(\gamma)$ and $(\beta)$ respectively. Let $O$ be the midpoint of the semicircle $(\alpha)$ on the opposite side of the arbelos. If the incircle of the arbelos is tangent to the semicircles $(\gamma)$ and $(\beta)$ and $(\alpha)$ at $P_2, P_1 , P_0$ respectively, then
- $A, C, P_2, P_0$ lie on a circle, center $Q$;
- $B, C, P_1, P_0$ lie on a circle, center $P$;
- The points $A$, $B$, $P_2$ and $P_1$ lie on a circle, center $O$.

*Proof*:

- Let us assume that the arbelos and the Pappus circle are constructed. Let $D$ be the intersection of the semicircle $(\alpha)$ with the line perpendicular to $(AB)$ at $C$. See (Figure ...).
 
  Note that $AB \cdot AC = AD^2$ ($[DC]$ height of the right triangle $ADB$ in $D$). 

  Consider the inversion $i_A$ with respect to the circle $\mathcal{C}(A,D)$ of center $A$ that goes through $D$.

  $i(B)=C$, and $i_A$ leaves the line $AB$ invariant. The semicircles $(\beta)$ and $(\alpha)$ have as respective inverses the half-straight lines  ${\ell}'$ and ${\ell}$ perpendicular to $(AB)$ at $C$ and $B$ respectively. The semicircle $\gamma$ is also invariant by $i_A$ because its center belongs to the line $(AB)$. The incircle $(\delta)$ or ($P_0P_1P_2$) of the arbelos is inverted into a circle tangent to the semicircle $(\gamma)$, and the lines $\ell$, (${\ell}'$), at $Q$ , $P_1'$, $P_0'$ respectively. 
 

  Since the semicircle $(\gamma)$ is invariant, the points $A$, $P_2$, and $Q$ are aligned because the image of $Q$ is $P_2$. 

  $CBP_1'P_0'$ is a square and $Q$ is its center thus $B$, $Q$ and $P_0'$ are aligned.

  Their inverses are therefore on a circle through $A$. $A$, $C$, $P_2$ and $P_0$ are concyclic.

  Let $O$ be the midpoint of the arc $\overset{\frown}{AB}$. $O, B, Q, P_0'$ are aligned. The tangent at $A$ to the circle $\mathcal{C}(A,D)$ is parallel to the line $(BL)$ so the diameter coming from $A$ is perpendicular to the line $(BL)$.
  The center of the circle $\mathcal{C}(A,D)$ belongs to the line $(AL)$ and to the perpendicular bisector of $[AC]$ so it is the point $P$.
- The same reasoning  shows that $B, C, P_1, P_0$ lie on the circle $\mathcal{C}(Q,B)$ with the inversion $i_B$ center $B$ through $D$.
- Since $C, Q, P_1'$ are on a line making a $45^\circ$ angle with $(AB)$, its inversive image by $i$ is a circle through $A, B, P_2, P_1$  also making a $45^\circ$ angle with $(AB)$. The center of this circle is necessarily the midpoint $O$ of the semicircle $(\alpha)$ on the opposite side of the arbelos.
$L$ belongs to the bisector of $[AB]$ and $\widehat{BAO} = 45^\circ$

  Join $A, O$ to intersect the line $\ell$ at $O'$. Since $\widehat{BAO'} = 45^\circ = \widehat{BP_0'O'}$, the four points $A, P_0', B, O'$ are concyclic. Considering the inversive image of the circle, we conclude that the line $(CP_0)$ passes through $O$.

  The center of the incircle can now be constructed as the intersection of the lines joining $P_2, P_1, P_0$ to the centers of the corresponding semicircles of the arbelos.

*Proof*: Consider the inversion $i_A$ with respect to the circle $\mathcal{C}(A,D)$ of center $A$ that goes through $D$.
The circle $(\gamma)$ is invariant by $i_A$ so $i_A(P_2) = Q$
  and  $A$, $P_2$, $Q$ are collinear, as are $B$, $P_1$ , $P$.  In Figure \ref{cocyclicpts}, let $T$ be the intersection of the line $(AQ)$ with the circle $(\delta)$. The inversive image $T'$ (in the circle $\mathcal{C}(A,D)$) is the intersection of the same line with the circle $QP_1'P_0'$. Note that. $\widehat{AT'P_0'} = \widehat{QT'P_0'}= \widehat{QP_1'P_0'}= 45^\circ = \widehat{ABP_0'}$ so that $A$, $B$, $T'$,$ P_0'$ are concyclic.

Considering the inversive image of this circle, we conclude that the line $(CP_0)$ contains $T$. In other words, the lines $(AQ)$ and $(CP_0)$ intersect at the point $T$ on the circle ($P_0P_1P_2$). Likewise, $(BQ)$ and $(CP_0)$ intersect at the same point.

We have another construction of the first Pappus circle.

- Construct $P$ and $Q$ the midpoints of the arcs $\overset{\frown}{AC}$ and $\overset{\frown}{CB}$;
