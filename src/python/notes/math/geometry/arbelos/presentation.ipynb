### Arbelos Definition by Archimedes' Book of Lemmas
If\: $[AB]$ be the diameter of a semicircle and C any point on $[AB]$, and if semicircles be described within the first semicircle and having $[AC]$, $[CB]$  as diameters respectively, the figure included between the circumferences of
  the three semicircles is  called *arbelos*. Let $\beta, \gamma, \alpha$ be the semicircles of respectives diameters $[AC]$, $[CB]$ and $[AB]$, then the arbelos bounded by the three semicircles is denoted $(\beta,\gamma,\alpha)$ or $(A,C,B)$.

In the rest of the document, we will refer to $a$, $b$ et $a+b$  the respective radii of the semicircles $\beta$, $\gamma$ and $\alpha$. $O_1$, $O_2$ and $O_0$ will be the respective centers of the semicircles.

### About the *Devine Proportion*
In mathematics, expressed algebraically, for quantities $a$ and $b$ with $a>b>0$,
\begin{equation*}
  \frac{a+b}{a} = \frac{a}{b} = \varphi
\end{equation*}
where the Greek letter $\varphi$  represents the *golden ratio* or *devine proportion*. In other words, two quantities are in the golden ratio if their ratio is the same as the ratio of their sum to the larger of the two quantities.

In the package *tkz-euclide*, you can use the macro *\tkzPhi* to get the value of golden ratio: $\varphi\approx 1.618034$.

We saw above that the Golden Ratio has the property:
\begin{equation*}
  \frac{a+b}{a} = \frac{a}{b}
\end{equation*}

We can split the first fraction

\begin{equation*}
  \frac{a}{a} + \frac{b}{a} = \frac{a}{b}
\end{equation*}

Either $1 +\dfrac{1}{\varphi} = \varphi $ with  $ \varphi = \dfrac{a}{b}$

Also $\varphi$ is a positive solution of :
\begin{equation*}
  x^2 - x-1 = 0
\end{equation*}

That is to say $\varphi=\dfrac{1+\sqrt{5}}{2} $
% subsubsection definition (end)

###  Algebraic Properties

- Powers of $\varphi$
  \begin{array}{ccc}
      \hline
      \varphi^2 & = &\varphi + 1 \\
      \varphi^3 & = &\varphi^2 + \varphi =2\varphi+1\\
      \varphi^4 & = &3\varphi+2\\
      \varphi^n & = &\varphi^{(n-1)}+\varphi^{(n-2)}\\
         \hline
  \end{array}
- Powers of $\dfrac{1}{\varphi}$
  \begin{array}{ccc}
      \hline
      \dfrac{1}{\varphi}   & =  &\varphi -1 \\
      \dfrac{1}{\varphi^2} & = &2 - \varphi = 1 - \dfrac{1}{\varphi} \\
      \dfrac{1}{\varphi^3} & = &2\varphi - 3 = \dfrac{2}{\varphi} -1\\
         \hline
  \end{array}

#### *Golden section of a line segment*
This is the simplest example of the golden ratio in relation to geometry. It's a special line segment portion of a line segment, called the golden segment,  illustrated here:


````python
%%tikz -f png -s 600,300 -p tkz-euclide,tkzexample,pgfornament -l math,calc,fpu
\tkzSetUpColors[background=white,text=darkgray]  
\tkzSetUpPoint[size=2,color=teal]
\tkzSetUpLine[ultra thin,color=teal]
\tkzSetUpCompass[color=orange,ultra thin,/tkzcompass/delta=10] 

\tkzDefPoints{0/0/A,10/0/B}
\tkzDefGoldenRatio(A,B)             \tkzGetPoint{C}
\tkzDrawSegment[teal](A,B)
\tkzDrawSegment[purple](C,B)
\tkzDrawPoints(A,B,C)
\tkzLabelPoints(A,B,C)
````


  We consider the segment $[AB]$ and we decide to define a point $C$, situated between $A$ and $B$ in order to have:
  \begin{equation*}
    \frac{AC}{CB} = \frac{AB}{AC}
  \end{equation*}

  The point $C$ defines a devine proportion on the segment $[AB]$. $AC$ is the golden section of the line segment $[AB]$.

  It's easy to get $C$ Just use the macro of \tkzname{tkz-euclide}: \tkzcname{tkzDefGoldenRatio(A,B)} \tkzcname{tkzGetPoint\{C\}}.


  With $AC=2a, CB=2b \quad \text{a>b}$


  We transformed this relation by doing a cross product :

  $AC^2 = AB \cdot CB$

  which becomes

  $4a^2 = 2(a+b) \cdot 2b $

  That we can write:

   $a^2=ab+b^2$ or $\dfrac{a}{a} + \dfrac{b}{a} = \dfrac{a}{b}$

- Obtaining the divine proportion with a ruler and a compass.

````python
%%tikz -f png -s 600,400 -p tkz-euclide,tkzexample,pgfornament -l math,calc,fpu -po rotate=-90
\tkzSetUpColors[background=white,text=darkgray]  
\tkzSetUpPoint[size=2,color=teal]
\tkzSetUpLine[ultra thin,color=teal]
\tkzSetUpCompass[color=orange,ultra thin,/tkzcompass/delta=10] 
\tikzset{label style/.append style={below,color=teal,font=\scriptsize}}
\tikzset{new/.style={color=orange,ultra thin}} 
\tikzset{step 1/.style={color=cyan,ultra thin}} 
\tikzset{step 2/.style={color=purple,ultra thin}}

\tkzDefPoints{0/0/A,5/0/B}
\tkzDefSquare(A,B)                \tkzGetPoints{E}{F}
\tkzDefMidPoint(B,E)              \tkzGetPoint{I}
\tkzInterLC[near](B,E)(I,F)       \tkzGetFirstPoint{J}
\tkzInterLL(F,J)(A,B)             \tkzGetPoint{C}
\tkzDrawPolygon(A,B,E,F)
\tkzDrawPoints(A,B,C,E,F,I,J)
\tkzDrawSegments[color=orange,dashed](F,J I,A B,J)
\tkzDrawArc[delta=10,dashed](I,A)(J)
\tkzLabelPoints(A,B,C,E,F,I,J)
\tkzMarkSegments[mark=s||](B,I I,E)
````

-
  - First we construct a square $ABEF$;
  - Then we construct the midpoint $I$ of segment $[BE]$;
  - The arc of center F passing through $I$ intersects the half-line $[AF)$ at $J$;
  - Finally the parallel to $(IJ)$ passing through $F$ intersects the segment $[AB]$ at $C$.

  Proof:  $[AC]$ is the golden segment of $[AB]$

  If $AF=AB=x$ then $AI=IJ=\sqrt{x^2+\dfrac{x^2}{4}} = \dfrac{x\sqrt{5}}{2}$

  $BJ=IJ - IB = \dfrac{x\sqrt{5}}{2} - \dfrac{x}{2} =\dfrac{ -1 + \sqrt{5}}{2}x=\dfrac{x}{\varphi} $

  Finally
  
  $$
    \frac{AC}{CB} =  \frac{AF}{BJ} = \varphi
  $$

  Remark: $\dfrac{EJ}{EB}=\varphi $


#### Divide a segment according to the divine proportion  with a ruler and a compass.

````python
%%tikz -f png -s 600,400 -p tkz-euclide,tkzexample,pgfornament -l math,calc,fpu
\tkzSetUpColors[background=white,text=darkgray]  
\tkzSetUpPoint[size=2,color=teal]
\tkzSetUpLine[ultra thin,color=teal]
\tkzSetUpCompass[color=orange,ultra thin,/tkzcompass/delta=10] 
\tikzset{label style/.append style={below,color=teal,font=\scriptsize}}
\tikzset{new/.style={color=orange,ultra thin}} 
\tikzset{step 1/.style={color=cyan,ultra thin}} 
\tikzset{step 2/.style={color=purple,ultra thin}}

  \tkzDefPoints{0/0/A,2.5/0/B}
  \tkzInterCC(A,B)(B,A)                  \tkzGetPoints{F}{E}
  \tkzInterLC[common=B](A,B)(B,A)        \tkzGetFirstPoint{I}
  \tkzDefMidPoint(A,B)                   \tkzGetPoint{J}
  \tkzDefMidPoint(J,B)                   \tkzGetPoint{K}
  \tkzDefPointWith[orthogonal](K,B)      \tkzGetPoint{k} 
  \tkzInterLC(K,k)(B,A)                  \tkzGetFirstPoint{G}
  \tkzInterLL(A,B)(E,G)                  \tkzGetPoint{C}
  \tkzDefPointBy[symmetry= center A](B)  \tkzGetPoint{O}
  \tkzDrawArc[delta=5](O,B)(G)
  \tkzDrawCircles(A,B B,A)
  \tkzDrawSegments(O,I A,E B,E)
  \tkzDrawSegments[purple](J,E A,G G,I K,G E,G)
  \tkzMarkSegments[mark=s||](A,E B,E O,A)
  \tkzDrawPoints(A,B,C,E,I,G,K,J,O)
  \tkzLabelPoints(A,B,C,E,I,G,K,J,O)
````

  - Let $J$ and $K$ be such that $AJ =\dfrac{AB}{2} $ and $JK =\dfrac{AB}{4} $
  - $GK^2 = AK\times KI$ then  $GK^2 =  \dfrac{3}{4}AB\times \dfrac{5}{4}AB$ then $GK = \dfrac{\sqrt{15}}{4}AB$
  - $JE =\dfrac{\sqrt{3}}{2}AB$ so $\dfrac{JE}{GK} = \dfrac{4\sqrt{3}}{2\sqrt{15}} =  \dfrac{2}{\sqrt{5}}$
  - $\dfrac{JC}{CK} = \dfrac{JE}{GK}$ and therefore $JC =  \dfrac{2}{\sqrt{5}} \times (JK-JC)$.
    It follows that $JC  = \dfrac{2}{\sqrt{5}+2}JK  = \dfrac{(\sqrt{5}-2)}{2}AB$
  - $AC = AJ + JC = \dfrac{AB}{2} + \dfrac{(\sqrt{5}-2)}{2}AB= \dfrac{(\sqrt{5}-1)}{2}AB =  \dfrac{2}{\sqrt{5}+1)}AB$
  - Finally $\dfrac{AB}{AC} = \varphi$

  Remark: It is possible to obtain the point $G$ by drawing the circle of center $O$ passing through $B$.

## Golden Arbelos

**Definition**: Let $\beta, \gamma, \alpha$ be the semicircles with the respective radii $a, b, a+b$. If the positive numbers $a$ and $b$ ($a>b$)  are in a \tkzname{golden ratio} in other words if
\begin{equation*}
  \frac{a+b}{a} = \frac{a}{b} = \varphi
\end{equation*}
then this special type of arbelos is called golden arbelos.


````python
%%tikz -f png -s 600,400 -p tkz-euclide,tkzexample,pgfornament,amsmath -l math,calc,fpu
\tkzSetUpColors[background=white,text=darkgray]  
\tkzSetUpPoint[size=2,color=teal]
\tkzSetUpLine[ultra thin,color=teal]
\tkzSetUpCompass[color=orange,ultra thin,/tkzcompass/delta=10] 
\tikzset{label style/.append style={below,color=teal,font=\scriptsize}}
\tikzset{new/.style={color=orange,ultra thin}} 
\tikzset{step 1/.style={color=cyan,ultra thin}} 
\tikzset{step 2/.style={color=purple,ultra thin}}

\tkzDefPoints{0/0/A,10/0/B}
\tkzDefGoldenRatio(A,B) \tkzGetPoint{C}
\tkzDefMidPoint(A,B)                     \tkzGetPoint{O_0}
\tkzDefMidPoint(A,C)                     \tkzGetPoint{O_1}
\tkzDefMidPoint(C,B)                     \tkzGetPoint{O_2}
\tkzLabelPoints(A,B,C)
\tkzDrawSegment(A,B)
\tkzDrawPoints(A,B,C)
\begin{scope}[local bounding box = graph]
  \tkzDrawSemiCircles[color=black](O_0,B)
\end{scope}
\useasboundingbox (graph.south west) rectangle (graph.north east);
\tkzClipCircle[out](O_1,C)\tkzClipCircle[out](O_2,B)
\tkzDrawSemiCircles[draw=none,fill=teal!15](O_0,B)
\tkzDrawSemiCircles[color=black](O_1,C O_2,B)
````

### First consequences

````python
%%tikz -f png -s 600,400 -p tkz-euclide,tkzexample,pgfornament,amsmath -l math,calc,fpu
\tkzSetUpColors[background=white,text=darkgray]  
\tkzSetUpPoint[size=2,color=teal]
\tkzSetUpLine[ultra thin,color=teal]
\tkzSetUpCompass[color=orange,ultra thin,/tkzcompass/delta=10] 
\tikzset{label style/.append style={below,color=teal,font=\scriptsize}}
\tikzset{new/.style={color=orange,ultra thin}} 
\tikzset{step 1/.style={color=cyan,ultra thin}} 
\tikzset{step 2/.style={color=purple,ultra thin}}

\tkzDefPoints{0/0/A,10/0/B}
\tkzDefGoldenRatio(A,B) \tkzGetPoint{C}
\tkzDefMidPoint(A,B)                               \tkzGetPoint{O_0}
\tkzDefMidPoint(A,C)                               \tkzGetPoint{O_1}
\tkzDefMidPoint(B,C)                               \tkzGetPoint{O_2}
\tkzDrawSemiCircles(O_0,B O_1,C O_2,B) 
\tkzDefPointBy[rotation=center C angle 90](B)      \tkzGetPoint{b}
\tkzInterLC[near](b,C)(O_0,A)                      \tkzGetFirstPoint{D}
\tkzInterCC(D,C)(O_0,B)                            \tkzGetPoints{U}{V}
\tkzDefPointBy[projection=onto U--V](O_1)          \tkzGetPoint{R}
\tkzDefPointBy[projection=onto U--V](O_2)          \tkzGetPoint{S}
\tkzDrawSegments(A,B A,D D,C D,B)
\tkzDrawPoints(A,B,C,D,R,S)
\tkzLabelPoints(A,B,C,R,S)
\tkzLabelPoints[above](D)
\tkzMarkRightAngles[fill=gray!20,opacity=.3](A,D,B D,C,A)
````

It is quite easy with the help of similar right-angled triangles to establish :

$BD = AC = 2a$, $CD =  2\sqrt{ab}$ and  $AD = 2a\sqrt{\varphi}$

$\dfrac{AD}{AT}=\dfrac{BD}{DW}=\varphi$
