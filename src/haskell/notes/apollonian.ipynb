## Apollonian Gasket

```haskell
import Diagrams.Prelude
import Diagrams.TwoD.Apollonian
import Data.Colour.Palette.BrewerSet
```

```haskell
purple :: [Colour Double]
purple = brewerSet Purples 9
```

```haskell
:set -XTypeFamilies

import Data.Colour (withOpacity)

blues   = map (blue `withOpacity`) [0.1, 0.2 .. 1.0]
alphaEx = hcat' (with & catMethod .~ Distrib & sep .~ 1 )
                (zipWith fcA blues (repeat (circle 1)))

colors  = brewerSet Pastel1 9
paletteEx = hsep 0.3 (zipWith fc colors (repeat (rect 0.5 1 # lw none)))
```

```haskell
diagram $ vsep 1 ([alphaEx, paletteEx] # map centerX) # bg black
```

```haskell
diagram $ alphaEx # bg black
```

```haskell
import Diagrams.Color.XKCD hiding (black, pink, red, lightblue, orange)

colors = [booger, poisonGreen, cinnamon, petrol, vibrantPurple]
```

```haskell
diagram $ hcat (map (`fcA` (circle 1 # lw none)) colors) # bg black
```

```haskell
s c     = square 1 # fc c
reds    = (s darkred ||| s red) === (s pink ||| s indianred)
diagram (hsep 1.5 . take 4 . iterate (opacity 0.7) $ reds) # bg black
```

```haskell
> cir = circle 1 # lw none # fc red
> overlap = (cir <> cir # translateX 1)
>
> example = hsep 1 [ overlap # opacity 0.3, overlap # opacityGroup 0.3 ]
>           # centerX
>        <> rect 9 0.1 # fc lightblue # lw none
```

```haskell
diagram $ example # bg black
```

```haskell
t = regPoly 3 1

example = hsep 0.2 [t, t # bg orange, t # bgFrame 0.1 orange]
```

```haskell
diagram $ example # bg black
```
